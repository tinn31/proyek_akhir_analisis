```python
def evaluate_dataframe(df, df_name):
    print(f"\n===== Menilai Tabel: {df_name} =====")

    print("\nInfo DataFrame:")
    df.info()

    print("\nJumlah Nilai yang Hilang (NaN):")
    print(df.isna().sum())

    print(f"\nJumlah Duplikasi: {df.duplicated().sum()}")

    print("\nStatistik Deskriptif:")
    print(df.describe())

for file in files:

    df = pd.read_csv(file)

    df_name = file.split('_')[2]

    evaluate_dataframe(df, df_name)
```

### Cleaning Data

```python
def clean_dataframe(df, df_name):
    print(f"\n===== Cleaning Tabel: {df_name} =====")

    before_duplicates = df.shape[0]
    df = df.drop_duplicates()
    after_duplicates = df.shape[0]
    print(f"Jumlah baris duplikat dihapus: {before_duplicates - after_duplicates}")

    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

    categorical_columns = df.select_dtypes(include=['object']).columns
    for col in categorical_columns:
        df[col] = df[col].fillna(df[col].mode()[0])

    print("\nJumlah Nilai yang Hilang setelah cleaning:")
    print(df.isna().sum())

    return df

cleaned_dfs = {}
all_cleaned_dataframes = []

for file in files:
    df = pd.read_csv(file)
    df_name = file.split('_')[2]
    cleaned_df = clean_dataframe(df, df_name)
    cleaned_dfs[df_name] = cleaned_df
    all_cleaned_dataframes.append(cleaned_df)
    cleaned_df.to_csv(f"cleaned_{df_name}.csv", index=False)


combined_df = pd.concat(all_cleaned_dataframes, ignore_index=True)
```

```python
for df_name, cleaned_df in cleaned_dfs.items():
    print(f"\n===== 5 Data Teratas dari Tabel: {df_name} =====")
    print(cleaned_df.head())
    print(f"===== 5 Data Terakhir dari Tabel: {df_name} =====")
    print(cleaned_df.tail())

print("\n===== 5 Data Teratas dari Tabel Gabungan =====")
print(combined_df.head())
print("===== 5 Data Terakhir dari Tabel Gabungan =====")
print(combined_df.tail())
```

### rata-rata konsentrasi polutan berdasarkan lokasi dan bulan.

```python

grouped_by_station_month = combined_df.groupby(['station', 'year', 'month'])['PM2.5'].mean()

print("\n===== Rata-rata Konsentrasi PM2.5 Berdasarkan Stasiun dan Bulan =====")
print(grouped_by_station_month)
```

 ### bandingkan rata-rata konsentrasi PM2.5 antar stasiun.

```python
average_pm25_by_station = combined_df.groupby('station')['PM2.5'].mean()

print("\n===== Rata-rata Konsentrasi PM2.5 Berdasarkan Stasiun =====")
print(average_pm25_by_station)
```

```python
from scipy import stats
import itertools


months = combined_df['month'].unique()

results = []

for month1, month2 in itertools.combinations(months, 2):

    data_month1 = combined_df[combined_df['month'] == month1]['PM2.5']
    data_month2 = combined_df[combined_df['month'] == month2]['PM2.5']

    t_stat, p_value = stats.ttest_ind(data_month1, data_month2)

    results.append({
        'Month 1': month1,
        'Month 2': month2,
        'T-statistic': t_stat,
        'P-value': p_value
    })


print("\n===== Hasil Uji T-test antara Semua Bulan =====")
for result in results:
    print(f"Antara Bulan {result['Month 1']} dan Bulan {result['Month 2']}: T-statistic = {result['T-statistic']:.4f}, P-value = {result['P-value']:.4f}")

    alpha = 0.05
    if result['P-value'] < alpha:
        print("Terdapat perbedaan signifikan.")
    else:
        print("Tidak terdapat perbedaan signifikan.")
```

### Pertanyaan 1: Bagaimana pengaruh variasi suhu (TEMP) terhadap konsentrasi PM2.5 dan PM10?

### Pertanyaan 2: Bagaimana tren konsentrasi PM2.5 dan PM10 sepanjang tahun?

```python
annual_trend = combined_df.groupby('year')[['PM2.5', 'PM10']].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='PM2.5', data=annual_trend, label='PM2.5', marker='o')
sns.lineplot(x='year', y='PM10', data=annual_trend, label='PM10', marker='o')
plt.title('Tren Konsentrasi PM2.5 dan PM10 Per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Konsentrasi')
plt.legend()
plt.show()
```

## Conclusion

### Pengaruh Variasi Suhu terhadap Konsentrasi PM2.5 dan PM10
Berdasarkan matriks korelasi yang diperoleh:

- PM2.5 dan PM10: Terdapat hubungan yang kuat antara konsentrasi PM2.5 dan PM10 dengan nilai korelasi 0.879. Ini menunjukkan bahwa ketika konsentrasi PM2.5 meningkat, konsentrasi PM10 juga cenderung meningkat.
-Suhu (TEMP): Terdapat korelasi negatif yang lemah antara suhu (TEMP) dengan kedua jenis partikel, yaitu -0.129 untuk PM2.5 dan -0.096 untuk PM10. Hal ini menunjukkan bahwa perubahan suhu tidak berpengaruh signifikan terhadap konsentrasi PM2.5 dan PM10. Dalam konteks ini, meskipun fluktuasi suhu terjadi, tidak ada bukti kuat bahwa suhu secara langsung mempengaruhi konsentrasi kedua jenis partikel tersebut.
